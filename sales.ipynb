```python
invoice_summary = df.groupby('InvoiceID').agg({
    'Date': 'first',
    'CustomerRef': 'first',
    'State': 'first',
    'type of state': 'first',
    'State_number': 'first',
    'InvoiceNetPrice': 'first',     
    'InvoiceItemPrice': 'sum',       
}).reset_index()

invoice_summary
```

```python
city_summary = invoice_summary[invoice_summary['type of state'] == 3]
sales_by_city = city_summary.groupby('State')['InvoiceNetPrice'].sum().sort_values(ascending=False)
sales_by_city
```

```python
top_city = sales_by_city.idxmax()
top_value = sales_by_city.max()

print(f'بیشترین فروش مربوط به شهر "{top_city}" با مبلغ {top_value:,.0f} تومان است.')
```
